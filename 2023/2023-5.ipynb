```python
from dataclasses import dataclass

minint = 0


@dataclass
class Mapping:
    src_start: int
    dst_start: int
    len: int

    @property
    def src_end(self):
        return self.src_start + self.len

    @property
    def dst_end(self):
        return self.dst_start + self.len


@dataclass
class Map:
    src: str
    dst: str
    mappings: list[Mapping]

    @property
    def src_starts(self) -> list[int]:
        transitions = []
        for mapping in self.mappings:
            transitions.append(mapping.src_start)
        return transitions

    @property
    def dst_starts(self) -> list[int]:
        transitions = []
        for mapping in self.mappings:
            transitions.append(mapping.dst_start)
        return transitions

    def src2dst(self, src: int) -> int:
        for mapping in self.mappings:
            if mapping.src_start <= src < mapping.src_start + mapping.len:
                return mapping.dst_start + (src - mapping.src_start)
        return src

    def dst2src(self, dst: int) -> int:
        for mapping in self.mappings:
            if mapping.dst_start <= dst < mapping.dst_start + mapping.len:
                return mapping.src_start + (dst - mapping.dst_start)
        return dst

    def __repr__(self) -> str:
        return f"{self.src} -> {self.dst}"


@dataclass
class Seed:
    id: int
    data: dict[str, int]

    def generate_data(
        self, maps: list[Map], _key: str | None = None, _id: int | None = None
    ) -> None:
        _key = _key or "seed"
        _id = _id or self.id

        for map in maps:
            if map.src == _key:
                break
        else:
            return

        self.data[map.dst] = map.src2dst(_id)
        self.generate_data(maps, map.dst, map.src2dst(_id))


# generate initial maps
maps: list[Map] = []
for section in puzzle.split("\n\n")[1:]:
    lines = section.splitlines()
    header = lines[0]
    src, dst = header.removesuffix(" map:").split("-to-")
    map = Map(src, dst, [])

    for line in lines[1:]:
        dst_start, src_start, length = [int(x) for x in line.split()]
        map.mappings.append(Mapping(src_start, dst_start, length))

    maps.append(map)

# fill up maps
for map in maps:
    if not any(mapping.src_start <= 0 for mapping in map.mappings):
        map.mappings.append(Mapping(0, 0, 1))

# generate seeds
seeds = []
for id in puzzle.splitlines()[0].split(":")[1].split():
    seed = Seed(int(id), {})
    seed.generate_data(maps)
    seeds.append(seed)

# find lowest location
min(seed.data["location"] for seed in seeds)

# seeds[0].data
```
