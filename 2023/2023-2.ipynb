```python
import re
from collections import namedtuple

Batch = namedtuple("Batch", ["red", "green", "blue"])


# parse...
data = {}

for line in puzzle.split("\n"):
    regex = re.search(r"Game (\d+):(.*)", line)
    assert regex

    key = int(regex.group(1))
    content = regex.group(2)
    batches_content = content.split(";")

    batches = []
    for batch_content in batches_content:
        batch = Batch(0, 0, 0)
        for color_count in batch_content.split(","):
            color_count = color_count.strip()
            regex = re.search(r"(\d+) (\w+)", color_count)
            assert regex
            count = int(regex.group(1))
            color = regex.group(2)
            if color == "red":
                batch = batch._replace(red=count)
            elif color == "green":
                batch = batch._replace(green=count)
            else:
                batch = batch._replace(blue=count)
        batches.append(batch)

    data[key] = batches

# assumptions:
max_red = 12
max_green = 13
max_blue = 14

valid_data = {}

for key, batches in data.items():
    if all(
        (
            batch.red <= max_red
            and batch.green <= max_green
            and batch.blue <= max_blue
        )
        for batch in batches
    ):
        valid_data[key] = batches

sum(valid_data.keys())
```

```python
powers = 0

for key, batches in data.items():
    reds = [batch.red for batch in batches]
    greens = [batch.green for batch in batches]
    blues = [batch.blue for batch in batches]

    power = max(reds) * max(greens) * max(blues)

    powers += power

powers
```
